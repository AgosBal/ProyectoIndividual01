```python
#Dado que en un primer intento de utilizar la librería autoviz, obtuve error en la columna 'Budget' por tipos de datos diferentes. Opté por convertirlo todo a valor numérico.

movies_completo['Budget'] = pd.to_numeric(movies_completo['Budget'], errors='coerce')

#Misma situación con 'Popularity'

movies_completo['Popularity'] = pd.to_numeric(movies_completo['Popularity'], errors='coerce')

#'Revenue'

movies_completo['Revenue'] = pd.to_numeric(movies_completo['Revenue'], errors='coerce')

#'Runtime'

movies_completo['Runtime'] = pd.to_numeric(movies_completo['Runtime'], errors='coerce')

#'VoteAverage'

movies_completo['VoteAverage'] = pd.to_numeric(movies_completo['VoteAverage'], errors='coerce')


#'VoteCount'

movies_completo['VoteCount'] = pd.to_numeric(movies_completo['VoteCount'], errors='coerce')

#'ReleaseYear'

movies_completo['ReleaseYear'] = pd.to_numeric(movies_completo['ReleaseYear'], errors='coerce')

#'ReleaseMonth'

movies_completo['ReleaseMonth'] = pd.to_numeric(movies_completo['ReleaseMonth'], errors='coerce')


#'Return'
movies_completo['Return'] = pd.to_numeric(movies_completo['Return'], errors='coerce')
```
